# import library beautifulsoup

```python
import requests
import urllib.request
from bs4 import BeautifulSoup
```

# request website

```python
response = requests.get('https://www.bbc.com/news/entertainment_and_arts')
doc = BeautifulSoup(response.text, 'html.parser')
```

# mengecek judul berita

```python
headlines = doc.find_all('h3')

for headline in headlines:
    print(headline.text)
```

# mencari link dalam website

```python
links = doc.find_all('a')

for link in links:
    print(link.text)
    print(link['href'])
```

# Memfilter link berita yang berhubungan dengan topic

```python
list_key_link = ['/news/entertainment-arts-']
berita_link = []
for i in list_key_link:
    berita_link = berita_link + [link for link in links if i in str(link) and 'javascript:void(0)' not in str(link) 
               and '#' not in str(link) and "video" not in str(link) and "Read more" not in str(link) and ":" not in str(link)]
print(len(berita_link))
for link in berita_link:
    print('%s: %s'%(link.text.strip(),str("https://www.bbc.com"+link["href"])))
```

# fungsi penuh untuk scrapping web

```python
def getBerita(link):
    html = urllib.request.urlopen(link).read()
    soup = BeautifulSoup(html, "lxml")

    judul = soup.find("h1","story-body__h1").text.strip()
    isi_berita=""

    table = soup.find('div','story-body__inner') 

    for row in table.findAll('p'):
        if "Follow us" in row.text:
            pass
        else:
            isi_berita+=row.text
    
    tanggal = soup.find("div","date").text 
    return [judul,isi_berita,tanggal]
```
